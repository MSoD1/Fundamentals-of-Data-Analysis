### Exercise 1:
*Create box plots on a single set of axes for all five experiments in the morley data set.*

```python
# Load Dataset
data = pd.read_csv("morley.csv")
```

**Note:** I used a few different methods to create a boxplot. The first approach looks at transforming the csv into a txt file, removing the header row and columns are separated by a comma. The data is the separated into 5 variables, each representing an experiment. The five variables are then grouped together in a boxplot. </br>

```python
array = np.loadtxt('morley.csv', skiprows = 1,delimiter=',')
```

```python
exper_1 = array[:20, 2]
exper_1
```

```python
exper_2 = array[20:40, 2]
exper_2
```

```python
exper_3 = array[40:60, 2]
exper_3
```

```python
# Select the data for the fourth experiment.
exper_4 = array[60:80, 2]
exper_4
```

```python
# Select the data for the fifth experiment.
exper_5 = array[80:100, 2]
exper_5
```

```python
# Create a box plot.
plt.boxplot(np.array([exper_1, exper_2]).T);
```

```python
plt.boxplot(np.array([exper_1, exper_2,exper_3,exper_4,exper_5]).T);
plt.xticks([1, 2, 3, 4,5])
plt.xlabel("Experiment")
plt.ylabel("Speed of Light (km/s minus 299,000)")
plt.show()
```

The second approach looks at the dataset as a whole, with experiments 1-5 on the x-axis and speed on the y-axis. The swarmplot function allows use to add jitters (scatterplot result) to the boxplot to understand data within the boxplot better.

```python
ax = sns.boxplot(x='Expt', y='Speed', data=data)
plt.xlabel("Experiment")
plt.ylabel("Speed of Light (km/s minus 299,000)")
```

```python
# Add jitter with the swarmplot function
ax = sns.boxplot(x='Expt', y='Speed', data=data)
ax = sns.swarmplot(x='Expt', y='Speed', data=data, color="grey")
plt.show()
```
