## Normal Distribution

#### What is Normal Distribution?

<div style="text-align: justify"> Commonly called normal distribution and formally called Gaussian distribution, it is one of the most easily distinguished distributions due to its bell-shaped curve. It is named after the German mathematician, Carl Friedrich Gauss, who developed a two-parameter exponential function in 1809. The probability distribution derives from variance. This variance, or uncertainty around an outcome, produces a probability distribution, which basically tells us what outcomes are relatively more likely and which outcomes are relatively less likely. It is the most important probability distribution in statistics because it accurately describes the distribution of values for many natural phenomena. Some common examples of datasets that follow Gaussian distribution are: </div>
1. Body temperature <br/>
2. Heights <br/>
3. IQ scores <br/>

#### Normal Distribution formula and Empirical Rule

Normal distribution is defined by the following probability density function: <br/>
<center> $ p(x) = \frac{exp \frac{-(x- \mu)^2}{2\sigma ^2}}{\sigma \sqrt{2\pi}}$ </center> <br/>
<div style="text-align: justify">  The exponential function e is a constant, μ is the mean, and σ is the standard deviation. <br/>
The empirical rule for normal distributions describes where most of the data in a normal distribution will appear, and it states the following:</div><br/>
 - 68.2% of the observations will appear within +/-1 standard deviation of the mean;<br/>
 - 95.4% of the observations will fall within +/-2 standard deviations; and<br/>
 - 99.7% of the observations will fall within +/-3 standard deviations.
 
<br/><div style="text-align: justify"> For example, in a normal distribution, 68% of the observations fall within +/- 1 standard deviation from the mean. This property is part of the Empirical Rule, which describes the percentage of the data that fall within specific numbers of standard deviations from the mean for bell-shaped curves. Any data points falling outside +/-3 is a rare occurance. <br/> </div>

#### Parameters of Normal Distribution

<div style="text-align: justify">  The following plot is an example of a normal distribution curve. The higher the blue line is in the plot, the higher the frequency of seeing that value below it on the x-axis. The value of 0 has a higher probability than the value of 3, this is because the mean is equal to 0 and the standard deviation in 1. Meaning the normally distributed random variable has a range of 1 to deviate from the mean of 0, while variables found at point 3 have deviated three standard deviations away from the mean 0. Many of these random variables occur between 1 and -1 because the standard deviation is 1 and the mean is 0. </div>

```python
# Package import
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
```

```python
#x-axis ranges from -3 and 3 with .001 steps
x = np.arange(-3, 3, 0.001)

#plot normal distribution with mean 0 and standard deviation 1
plt.plot(x, norm.pdf(x, 0, 1))
```

#### - Mean

<div style="text-align: justify"> As with all probability distributes, the parameters form the shape of the distribution. Changing the mean will change the location of the curve, a higher mean value will result in a curve found further along the x-axis even if the standard deviation is the same. The mean is the central tendency of the normal distribution while the standard deviation is the measure of variability. The below graph is an example of how changing the mean shifts the entire curve left or right on the X-axis.<div/>

```python
#x-axis ranges from -5 and 5 with .001 steps
x = np.arange(-5, 5, 0.001)

#define multiple normal distributions
plt.plot(x, norm.pdf(x, -2, 1), label='μ: -2, σ: 1')
plt.plot(x, norm.pdf(x, 0, 1), label='μ:0, σ: 1')
plt.plot(x, norm.pdf(x, 2, 1), label='μ:2, σ: 1')

#add legend to plot
plt.legend()
```

#### - Standard Deviation

<div style="text-align: justify"> The standard deviation determines how far the values tend to deviate from the mean. Graphically, changing the standard deviation would result in 
either tightening or spreading the bell-shaped curve along the x-axis. Smaller deviations result in tighter shaped curves because fewer values deviate 
from the mean, probabilities are higher that values won't fall far from the mean. If the standard deviation is increased, then the likelihood
that observations would be found further from the mean would also increase. <div/>

```python
#x-axis ranges from -5 and 5 with .001 steps
x = np.arange(-5, 5, 0.001)

#define multiple normal distributions
plt.plot(x, norm.pdf(x, 0, 1), label='μ: 0, σ: 1')
plt.plot(x, norm.pdf(x, 0, 1.5), label='μ:0, σ: 1.5')
plt.plot(x, norm.pdf(x, 0, 2), label='μ:0, σ: 2')

#add legend to plot
plt.legend()
```

#### Central Limit Theorem

<div style="text-align: justify"> The Central Limit Theorem states that the sampling distribution of the sample means from a given population will follow a normal distribution. In statistics there is a general rule that only populations larger than 30 will give accurate readings. As the population increases, the sampling distribution of the mean follows a normal distribution even when the underlying distribution of the original variable is non-normal. </div>

#### Skewness and Kurtosis

<div style="text-align: justify"> As previously discussed, with normal distribution curves most of the observations cluster around the central peak. Skewness represents a distribution's degree of symmetry. Because normal distributions result in perfectly symmetric curves, there exists zero skewness since mode=median=mean. If the observations are skewed, then a normal distribution cannot be used to model the data and a different distribution should be considered. <br/>
Kurtosis indicates how much data resides in the tails, the data points found furthest from the mean and either side of the curve. A distribution can be sharply peaked with low kurtosis, and a distribution can have a lower peak with high kurtosis. For a normal distribution, kurtosis is always equal to 3 else the data deviates from a normal distribution. <div/>
