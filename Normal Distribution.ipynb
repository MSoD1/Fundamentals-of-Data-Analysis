### Normal Distribution

Normal or Gaussian distribution is one of the most easily distinguished distribution due to its bell-shaped curve. 
The distribution is one of the most important probability distributions in statistics for independent, random variables.
It is important because it accurately describes the probability distribution of values for many natural phenomena. Common examples of
normal distributions are heights, blood pressure and IQ scores.

The curve of the distribution, which is a continuous probability distribution has two parameters, the mean and the standard deviation. 
The mean is the point at which the graph is symmetrically shaped around, while the standard deviation determines the dispersion from the mean. 
Most points cluster around the mean, which is the central peak or maximum, the probability of points further from the mean taper off equally in both directions.

The normal distribution is produced by the normal density function, $ p(x) = \frac{exp \frac{-(x- \mu)^2}{2\sigma ^2}}{\sigma \sqrt{2\pi}}$ <br/>
In this exponential function e is the constant 2.71828…, μ is the mean, and σ is the standard deviation. 
The probability of a random variable falling within any given range of values is equal to the proportion of the 
area enclosed under the function’s graph between the given values and above the x-axis.


```python
# Package import
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
```

```python
#x-axis ranges from -3 and 3 with .001 steps
x = np.arange(-3, 3, 0.001)

#plot normal distribution with mean 0 and standard deviation 1
plt.plot(x, norm.pdf(x, 0, 1))
```

As with all probability distributes, the parameters form the shape of the distribution. 
Changing the mean will change the location of the curve, the higher the mean value is found further along the x-axis even if the standard deviation is the same.
The mean is the central tendency of the normal distribution while the standard deviation is the measure of variability. 
The standard deviation determines how far the values tend to deviate from the mean. Graphically, changing the standard deviation would result in 
either tightening or spreading the bell-shaped curve along the x-axis. Smaller deviations result in tighter shaped curves because fewer values deviate 
from the mean, probabilities are higher that values won't fall far from the mean. If the standard deviation is increased, then the likelihood
that observations would be found further from the mean would also increase.

```python
#x-axis ranges from -5 and 5 with .001 steps
x = np.arange(-5, 5, 0.001)

#define multiple normal distributions
plt.plot(x, norm.pdf(x, 0, 1), label='μ: 0, σ: 1')
plt.plot(x, norm.pdf(x, 0, 1.5), label='μ:0, σ: 1.5')
plt.plot(x, norm.pdf(x, 0, 2), label='μ:0, σ: 2')

#add legend to plot
plt.legend()
```
