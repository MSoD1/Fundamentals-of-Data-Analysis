### Exercise 1: 
*If we only wanted exactly four heads as opssosed to five, the equivalent calculation would be (10 * 9 * 8 * 7)/(4 * 3 * 2 * 1) </br>
Does that evenly divide too? What is the formula in general? </br>
Does it always come out as a positive whole number?*

(10 * 9 * 8 * 7)/(4 * 3 * 2 * 1)

We refer to this as “N choose r.” Sometimes the number of combinations is known as a
binomial coefficient, and sometimes the notation NCr is used. So, in the present example, 
$$
\frac{N}{r}  = \frac{10}{4}  = \frac{N!}{r!(N-r)!}  = \frac{10!}{4!(10-4)!} = \frac{10 * 9 * 8 * 7}{4 * 3 * 2 * 1} = \frac{5040}{24} = 210
$$

Yes the result based on Pascal's Rule is always a positive integer
